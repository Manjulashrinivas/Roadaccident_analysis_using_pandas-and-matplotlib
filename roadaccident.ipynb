```python
#load csv file
import pandas as pd
df=pd.read_csv("indian_roads_dataset.csv")
print(df)
```

```python
print(df.columns)
```

```python
#accident count by state
accident_count_by_state=df.groupby("state")["accident_id"].count()
#print(accident_count_by_state)
plt.figure(figsize=(15,5))
accident_count_by_state_chart=plt.bar(accident_count_by_state.index,accident_count_by_state.values,color=plt.cm.Set2.colors)
labels=[f"{count}" for state,count in zip(accident_count_by_state.index,accident_count_by_state.values)]
plt.title("Statewise Accident Counts")
plt.xlabel("State")
plt.ylabel("Count")
plt.bar_label(accident_count_by_state_chart,labels=labels)
plt.show()

```

```python
#accident count by causes
accident_count_by_causes=df.groupby("cause")["accident_id"].count()
#print(accident_count_by_causes)
plt.figure(figsize=(15,5))
labels=[f"{cause}:{count}" for cause,count in zip(accident_count_by_causes.index,accident_count_by_causes.values)]
plt.pie(accident_count_by_causes.values,labels=labels,colors=plt.cm.Set2.colors,autopct="%1.1f%%")
plt.title("Reason For accident and Count")
plt.show()

```

```python
#accident_severity count
accident_severity_count=df.groupby("accident_severity")["accident_id"].count()
print(accident_severity_count)
plt.figure(figsize=(15,5))
accident_severity_count_chart=plt.barh(accident_severity_count.index,accident_severity_count.values,color=plt.cm.Set1.colors)
labels=[f"{accident_severity}:{count}"for accident_severity,count in zip(accident_severity_count.index,accident_severity_count.values)]
plt.title("accident_severity Counts")
plt.xlabel("Counts")
plt.ylabel("accident_severity")
plt.bar_label(accident_severity_count_chart,labels=labels)
plt.show()
```

```python
import matplotlib.pyplot as plt
df["is_weekend"]=df["is_weekend"].replace({1:"weekend",0:"Weekdays"})
#print(df["is_weekend"])

#weekend and weekdays accident count 
weekend_and_weekdays_accident_count=df.groupby("is_weekend")["accident_id"].count()
print(weekend_and_weekdays_accident_count)
plt.figure(figsize=(15,5))
labels=[f"{day}:{count}"for day,count in zip(weekend_and_weekdays_accident_count.index,weekend_and_weekdays_accident_count)]
plt.pie(weekend_and_weekdays_accident_count.values,labels=labels,colors=plt.cm.Set2.colors,autopct="%1.1f%%",wedgeprops={"width":0.4})
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# convert date to datetime
df["date"] = pd.to_datetime(df["date"])

# extract year
df["year"] = df["date"].dt.year
#print(df["year"])

# accident count per year
year_accident_count = df.groupby("year")["accident_id"].count()
print(year_accident_count)

# plot
plt.figure(figsize=(8,5))
plt.plot(year_accident_count.index, year_accident_count.values, marker="o")

plt.xlabel("Year")
plt.ylabel("Count")
plt.xticks(year_accident_count.index)

plt.title("Accident Rate Over the Year")

plt.show()
```
